# Creating the Course-Skill Dataframe
The excel sheet being imported is just a template with no important values (all values are set to 0) <br>

Data is inserted/changed in the dataframe using df.at[row_name, column_name] <br>

Whenever a new skill or class is included in an insert statement, a new row/column is created to house the value while the default value across the other rows/columns is 0 <br>

For the tech/soft column, a 0 represents a tech skill and a 1 represents a soft skill (because 0 is the default and there are much more tech skills listed than soft skills) <br>

For the course columns, a 0 represents that the class does not teach the skill in that row while a 1 represents that the class does teach the skill

# Collapsing the dataframe into a human-readable list
First make a dataframe with all the combinations between the skills and courses with a 1 as their value while still keeping the tech/soft skill information

```python
import numpy as np
```

```python
skillStack = df.copy().drop(['Tech/Soft'], axis=1)
skillStack = skillStack.where(skillStack != 1, skillStack.columns.to_series(), axis=1) #replaces the 1s in the columns with the column name
cols = skillStack.columns[0:]
skillStack[cols] = skillStack[cols].replace({'0':np.nan, 0:np.nan}) #replaces zeros to nan which will be dropped during the stack
stacked = skillStack.stack()
skillStack = stacked.reset_index(name='Course').drop(['level_1'], axis=1)
skillStack = skillStack.set_index('Skills').join(df.iloc[:,0:1]).replace({'0':'Technical', 0:'Technical', 1:'Soft', '1':'Soft'}) #Add back the Tech/Soft column according to index and replace the 0s with 'Technical' and 1s with 'Soft'
skillStack = skillStack.reset_index()
skillStack
```

## Collapse the Courses into a List of Courses
This dataframe shows the list of courses that teach each skill

```python
groupedSkills = skillStack.set_index('Skills').groupby('Skills')['Course'].apply(list).reset_index(name="Course").set_index('Skills').join(df.iloc[:,0:1]).replace({'0':'Technical', 0:'Technical', 1:'Soft', '1':'Soft'})
groupedSkills
```
