```python
import os
import pandas as pd

# Get the dataset path
path = kagglehub.dataset_download("joebeachcapital/linkedin-jobs")

# List the files in the directory
files = os.listdir(path)
print("Files in the directory:", files)

# Assuming there is a CSV file in the dataset, find and load it
csv_file = None
for file in files:
    if file.endswith(".csv"):
        csv_file = os.path.join(path, file)
        break

if csv_file:
    # Load the CSV into a pandas DataFrame
    df = pd.read_csv(csv_file)
    print("CSV loaded successfully. Here are the first few rows:")
    print(df.head())
else:
    print("No CSV file found in the dataset directory.")
```

```python
df.columns
```

```python
df.drop(['Class','Total_applicants','LinkedIn_Followers','Location'],axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
df.groupby('Involvement').size().plot(
    kind='pie',
    color=sns.color_palette('Dark2')
)


plt.gca().spines[['top', 'right']].set_visible(False)
plt.xlabel("Count", fontsize=14)
plt.ylabel("Involvement", fontsize=14)
plt.title("Company Involvement Distribution", fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
df.groupby('Industry').size().plot(
    kind='barh',
    color=sns.color_palette('Dark2')
)


plt.gca().spines[['top', 'right']].set_visible(False)
plt.xlabel("Count", fontsize=14)
plt.ylabel("Industry", fontsize=14)
plt.title("Company Industry Distribution", fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
keywords = [
    "analyst", "engineer", "business", "developer", "information", "technology",
    "dev", "data", "management", "manage", "manager", "manufacturing", "marketing",
    "sales", "scientist", "project", "project manager"
]



keyword_counts = {
    keyword: df['Designation'].str.contains(keyword, case=False, na=False).sum() for keyword in keywords
}

keyword_df = pd.DataFrame(list(keyword_counts.items()), columns=['Keyword', 'Count'])

# plot the bar graph
plt.figure(figsize=(14, 8))
plt.bar(keyword_df['Keyword'], keyword_df['Count'], color='skyblue')
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.title('Frequency of Keywords in Glassdoor Job Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
