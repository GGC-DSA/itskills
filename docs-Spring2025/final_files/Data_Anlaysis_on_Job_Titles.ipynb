# Now working with our own data

```python
glassdoor=pd.read_csv('glassdoor_cleaned.csv')
glassdoor.columns
```

```python
indeed=pd.read_csv('indeed_cleaned.csv')
indeed.columns
```

```python
linkedin=pd.read_csv('linkedin_cleaned.csv')
linkedin.columns
```

```python
indeed.rename(columns={'description':'job_summary'},inplace=True)
linkedin.rename(columns={'job_employment_type':'job_type'},inplace=True)
glassdoor.rename(columns={'job_overview':'job_summary'},inplace=True)
new_data_merged=pd.merge(linkedin,indeed,how='outer',on=['job_title','job_summary','job_type','company_name'])
new_data_merged=pd.merge(new_data_merged,glassdoor,how='outer',on=['job_title','job_summary','company_name'])
new_data_merged.head()
```

```python
enterprise_systems_keywords = [
    "ERP", "SAP", "Oracle", "Microsoft Dynamics", "NetSuite", "Business Systems", "Enterprise Architect",
    "CRM", "Workday", "PeopleSoft", "System Analyst", "Enterprise Engineer", "Business Applications",
    "Process Automation", "IT Operations", "Enterprise Integration", "IT Consultant", "Enterprise Solutions"
]

system_security_keywords = [
    "Cybersecurity", "Security", "Network Security", "System Security", "Penetration Testing",
    "SOC Analyst", "Information Security", "Threat Analysis", "Firewall", "Cloud Security",
    "Ethical Hacking", "IAM", "Risk Management", "Compliance", "Data Protection", "Incident Response",
    "IT Auditor", "Vulnerability", "Forensics", "Zero Trust", "Endpoint Security", "DevSecOps"
]

digital_media_keywords = [
    "Digital Media", "Content Creator", "Video Editor", "Graphic Designer", "Multimedia", "Animation",
    "Social Media", "Marketing", "SEO", "Adobe", "UX/UI Designer", "Web Designer", "Creative Director",
    "Branding", "Visual Design", "Content Marketing", "Social Media Manager", "User Experience",
    "Digital Marketing", "Motion Graphics", "Visual Effects", "Content Strategy", "Influencer", "Copywriting"
]
```

```python
def categorize_job_title(title):
    title = str(title).lower()  # Convert to lowercase

    if any(keyword.lower() in title for keyword in enterprise_systems_keywords):
        return "Enterprise Systems"
    elif any(keyword.lower() in title for keyword in system_security_keywords):
        return "System & Security"
    elif any(keyword.lower() in title for keyword in digital_media_keywords):
        return "Digital Media"
    else:
        return None

# Apply categorization
new_data_merged["job_category"] = new_data_merged["job_summary"].apply(categorize_job_title)

# Remove rows where job_category is None (i.e., not matching our categories)
new_data_merged_filtered = new_data_merged.dropna(subset=["job_category"])

# Display result
new_data_merged_filtered["job_category"].value_counts()
```

```python
es_keyword_counts = {
    keyword: new_data_merged_filtered[new_data_merged_filtered['job_category'] == 'Enterprise Systems']['job_summary']
    .str.contains(keyword, case=False, na=False)
    .sum()
    for keyword in enterprise_systems_keywords
}

es_keyword_df = pd.DataFrame(list(es_keyword_counts.items()), columns=['Keyword', 'Count'])

# Sort and display top 20
es_keyword_df = es_keyword_df.sort_values(by='Count', ascending=False)
print(es_keyword_df.head(20))
```

```python
system_security_keyword_counts = {
    keyword: new_data_merged_filtered[new_data_merged_filtered['job_category'] == 'System & Security']['job_summary']
    .str.contains(keyword, case=False, na=False)
    .sum()
    for keyword in system_security_keywords
}

system_security_keyword_df = pd.DataFrame(list(system_security_keyword_counts.items()), columns=['Keyword', 'Count'])

# Sort and display top 20
system_security_keyword_df = system_security_keyword_df.sort_values(by='Count', ascending=False)
print(system_security_keyword_df.head(20))
```

```python
dm_keyword_counts = {
    keyword: new_data_merged_filtered[new_data_merged_filtered['job_category'] == 'Digital Media']['job_summary']
    .str.contains(keyword, case=False, na=False)
    .sum()
    for keyword in digital_media_keywords
}

dm_keyword_df = pd.DataFrame(list(dm_keyword_counts.items()), columns=['Keyword', 'Count'])

# Sort and display top 20
dm_keyword_df = dm_keyword_df.sort_values(by='Count', ascending=False)
print(dm_keyword_df.head(20))
```
